```python
ticker = 'AJ'
url = F'https://classic.set.or.th/set/factsheet.do?symbol={ticker}&ssoPageId=3&language=en&country=TH'
```

```python
df = pd.read_html(url)
```

```python
def find_correct_index(list_of_data_frame):
    for number in range(1,7):
            if list_of_data_frame[number+14][0][0] == 'Accumulated: Turnover Ratio, Value Trade/Day':
                return number+14
```

```python
df[find_correct_index(df)]
```
